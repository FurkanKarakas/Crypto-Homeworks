```python
possibleprimes = [p for p in xrange(4*c2size+1) if p in Primes() and (p-1) % 16 == 0]
print(possibleprimes)
```

```python
i = 0
possible_a_values = []
for p in possibleprimes:
    possible_a_values.append([y for y in xrange(floor(p/2),ceil(4*p/7)+1) if y%2==0])
print(possible_a_values[-1])
```

```python
c2string = str(bin(c2))
c2string = c2string[2:]
i=0
j=1
passed=true
possible_p_and_a_values = []
for p in possibleprimes:
    for a in possible_a_values[i]:
        j=1
        passed=true
        for char in c2string:
            if mod(j,16) in xrange(8,16):
                j+=1
                continue
            temp=mod((j+1)*a,p).lift()%2
            if temp!=int(char):
                passed=false
                break
            j+=1
        if passed:
            possible_p_and_a_values.append((p,a))
    i+=1
```
